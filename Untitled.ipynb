```python
import pandas as pd
import numpy as np
from PIL import Image
import numpy as np
import os 
import numpy as np
from matplotlib import pyplot as plt
import cv2
 


# Read the image
image = Image.open('1.JPG')
numpydata = np.asarray(image)
print("Image shape: ",numpydata.shape)
xcoordinates = range(0,numpydata.shape[0])
xcoordinates = np.transpose([xcoordinates] *numpydata.shape[1])
xcoordinates = xcoordinates.reshape(1,xcoordinates.shape[0]*xcoordinates.shape[1])
ycoordinates = range(0,numpydata.shape[1])
ycoordinates = np.vstack([ycoordinates]*numpydata.shape[0])
ycoordinates = ycoordinates.reshape(1,ycoordinates.shape[0]*ycoordinates.shape[1])
values = numpydata.reshape(1,numpydata.shape[0]*numpydata.shape[1])
data = pd.DataFrame()
data['X'] =xcoordinates.tolist()[0]
data['Y'] =ycoordinates.tolist()[0]
data['Values'] =values.tolist()[0]
LATmin = 28 
LATmax = 63
LONmin = 8
LONmax = 37.8
lat = np.linspace(LATmin, LATmax, numpydata.shape[1])
lat = np.vstack([lat]*numpydata.shape[0])
lon = np.linspace(LONmax, LONmin, numpydata.shape[0]) 
lon = np.transpose([lon] *numpydata.shape[1])
x1 = lat.copy()
x2 = lat.copy()
x3 = lat.copy()
x4 = lat.copy()
xx1 = 51.56628497139077
xx2 =  56.60900958076577
x1[x1>=xx1] = 1
x2[x2<=xx2] = 1
x3[x3<xx1] = 0
x4[x4>xx2] = 0
X = np.multiply(np.multiply(np.multiply(np.multiply(x1, x2),x3),x4),lat)
yy1 = 22.621496549477314
yy2 = 26.37945060916716
x1 = lon.copy()
x2 = lon.copy()
x3 = lon.copy()
x4 = lon.copy()
x1[x1>=yy1] = 1
x2[x2<=yy2] = 1
x3[x3<yy1] = 0
x4[x4>yy2] = 0
Y = np.multiply(np.multiply(np.multiply(np.multiply(x1, x2),x3),x4),lat)
finalimage = np.multiply(np.multiply(X,Y),numpydata)
plt.imshow(finalimage)
plt.show()
```
